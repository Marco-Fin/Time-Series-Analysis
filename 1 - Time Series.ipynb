### Install Libraries

```python
from pandas import read_csv

import matplotlib.pylab as plt
import plotly.express as px
import yfinance as yf

plt.rcParams['figure.figsize']=(20,10)
plt.style.use('seaborn')
```

### Download Data

Our data are based on the shares of the company Petróleo Brasileiro S.A. — Petrobras (PETR4.SA).
https://finance.yahoo.com/quote/PETR4.SA

```python
series_petr4 = yf.download("PETR4.SA", start="2015-01-01", end="2020-10-02")
```

```python
series_petr4.head(10)
```

```python
series_petr4.columns
```

```python
import numpy as np
```

```python
series_petr4["AdjCloseRound"] = np.round(series_petr4['Adj Close'],2)
```

```python
series_petr4.head(10)
```

#### Time Series Visualization
#### OPEN

```python
series_petr4.plot(y="Open", figsize=(12,8), color="red")
plt.xlabel("Date", fontsize=18)
plt.ylabel("Open", fontsize=18)

plt.show()
```

#### Time Series Visualization
#### OPEN and CLOSE

```python
series_petr4[["Close", "Open"]].plot(figsize=(12,8))
plt.xlabel("Date", fontsize=18)
plt.ylabel("PETR4.SA", fontsize=18)
plt.show()
```
